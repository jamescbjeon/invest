```python
import yfinance as yf
import pandas as pd
```

```python
data = yf.download(['^GSPC']).Close.to_frame()
sp500 = data[data.index > '2000-01-01'].copy()

sp500
```

필요 데이터
* Max value
* Max 여부
* 경과일수
* Drawdown %

1. Close 오늘 > 전일 비교
    * 오늘 > 전일 : 오늘 값 --> Highest
    * 오늘 <= 전일 : 전일 값 --> Highest
2. Highest == 오늘 Close
    * == : 0 --> PERIOD
    * != : (전일 - 오늘) + 전일PERIOD = PERIOD
3. Drawdown %
    * (오늘 / Highest - 1) * 100

```python
sp500['SHIFT'] = sp500.Close.shift()
sp500.dropna(inplace= True)

sp500
```

```python
sp500['MAX'] = (sp500.Close > sp500.SHIFT)
sp500
```

```python
sp500.MAX.apply(lambda x: sp500.Close if x is True else sp500.SHIFT)
```
