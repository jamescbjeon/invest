```python
from datetime import datetime

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rc('font', family='AppleGothic')

# !pip install yfinance
import yfinance as yf
```

```python
etflist = pd.read_excel('etflist_2021.xlsx')
etflist.TICKER = etflist.TICKER.map('{}.KS'.format)

etflist
```

```python
today = datetime.today().strftime('%Y-%m-%d')

start = '2020-01-01'
end = today
```

```python
# Create a dataframe to store the adjustes close price of the stocks
df = pd.DataFrame()

# Store the adjusted close price of the stock into the df
for ticker in etflist.TICKER:
    df[ticker] = yf.download(ticker, start, end)['Adj Close']
```

```python
reference = pd.DataFrame()

for ticker in ['^KS11', '^GSPC']:
    reference[ticker] = yf.download(ticker, start, end)['Adj Close']
```

```python
ref = reference / reference.iloc[0]
```

```python
df.dropna(axis=1, inplace= True)
df.head()
```

```python
dfper = df / df.iloc[0]
dfper = pd.concat([ref, dfper], axis=1)
```

```python
op_2y = (dfper.iloc[-1] - dfper.iloc[0]) / dfper.iloc[0]
```

```python
perf2y = pd.DataFrame()
perf2y['Output_2Y'] = op_2y.sort_values(ascending= False)
```

```python
perf2y.reset_index(inplace= True)
perf2y.rename(columns={"index": "TICKER"}, inplace= True)
perf = perf2y.merge(etflist, how='left', on='TICKER')
```
