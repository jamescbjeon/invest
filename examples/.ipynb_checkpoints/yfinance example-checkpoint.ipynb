## 회사 기본 정보

### .info : 요약 정보

```python
aapl_yf = yf.Ticker('AAPL')
```

```python
# .info : 회사 관련 정보를 json 형태로 반환
aapl_info = aapl_yf.info

aapl_info
```

### .actions : 배당, 분할

```python
aapl_act = aapl_yf.actions

print(aapl_act)
# print(aapl_data.to_string())  # 전체내용 모두 보고싶을 때
```

```python
# 배당 이력만 별도 확인
aapl_div = aapl_yf.dividends
aapl_div
```

```python
# 물적분할 이력만 별도 확인
aapl_split = aapl_yf.splits
aapl_split
```
