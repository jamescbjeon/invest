```python
start = '2020-02-01'
end = '2022-02-28'

tck1 = '245340.KS'  # 245340.KS TIGER 미국다우존스30
tck2 = '195980.KS'  # 195980.KS ARIRANG 신흥국MSCI(합성H)
tck3 = '148070.KS'  # 148070.KS KOSEF 국고채10년
tck4 = '245340.KS'  # 153130.KS KODEX 단기채권

df1 = yf.download(tck1, start, end)
df2 = yf.download(tck2, start, end)
df3 = yf.download(tck3, start, end)
df4 = yf.download(tck4, start, end)
```
