```python
# Import libraries
from datetime import datetime

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'AppleGothic' # Matplotlib 한글서체 오류처리
plt.rc('axes', unicode_minus=False)

# !pip install yfinance
import yfinance as yf
```

```python
# Get the stock/portfolio starting date
StartDate = '2022-01-01'

# Get the stocks ending data (today)
today = datetime.today().strftime('%Y-%m-%d')

# 주요 벤치마크 지표와 Performance 비교
bmticker = {
    'S&P500': '^GSPC', 
    'Nasdaq': '^IXIC', 
    'KOSPI':'^KS11'
}
sectorticker = {
    'IT': 'XLK',
    'Commuication': 'XLC',
    'HealthCare': 'XLV',
    'Financials': 'XLF',
    'ConStaple': 'XLP',
    'ConDiscret': 'XLY',
    'Indusrials': 'XLI',
    'Energy': 'XLE',
    'Materials': 'XLB',
    'Utilities': 'XLU',
    'RealEstate' : 'IYR'
}

df = pd.DataFrame()

for tk in bmticker:
    tk_yahoo = yf.Ticker(bmticker[tk])
    data = tk_yahoo.history(start= StartDate)['Close']
    df[tk] = data
    
for tk in sectorticker:
    tk_yahoo = yf.Ticker(sectorticker[tk])
    data = tk_yahoo.history(start= StartDate)['Close']
    df[tk] = data
    
df = df.dropna()    
# df = df.dropna().astype('int32')

df = df / df.iloc[0] * 100
```

```python
for sector in sectorticker:
    df[['S&P500', sector]].plot(figsize= (12, 4))
```
